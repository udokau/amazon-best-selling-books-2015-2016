```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from time import sleep
import lxml
```

```python
url = 'https://www.amazon.com/gp/bestsellers/2015/books/ref=zg_bsar_cal_ye'

headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
               "AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36",
               "Accept-Encoding":"gzip, deflate",
               "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", 
               "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

response = requests.get(url,headers = headers)
sleep(10)
content = response.content
sleep(10)
soup =BeautifulSoup(content, "lxml")
```

```python
url2 = 'https://www.amazon.com/gp/bestsellers/2016/books/ref=zg_bsar_cal_ye'

headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
               "AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36",
               "Accept-Encoding":"gzip, deflate",
               "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", 
               "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

response2 = requests.get(url2,headers = headers)
sleep(10)
content2 = response2.content
sleep(10)
soup2 =BeautifulSoup(content2, "lxml")
```

```python
books = soup.select("#gridItemRoot")
```

```python
data= []
for item in books:
    
    title = item.find('div', class_ ="_cDEzb_p13n-sc-css-line-clamp-1_1Fn1y").text
    
    try:
        author = item.find('a', class_ ="a-size-small a-link-child").text
    except Exception as e:
        author = np.nan
        
    book_format = item.find('span', class_ ="a-size-small a-color-secondary a-text-normal").text
    
    review = item.select(".a-size-small")[2].text
    
    rating = item.find('span', class_="a-icon-alt").text.split(' ')[0]
    
    try:
        price = item.find('span', class_ ="_cDEzb_p13n-sc-price_3mJ9Z").text[1:]
    except Exception as e:
        price = np.nan
    
    try:
        year = item.find('h1',class_="a-size-large a-spacing-medium a-text-bold").text[-1]
    except Exception as e:
        year = '2015'
        
    data.append([title, author, book_format, review, rating, price, year])
```

```python
books2 = soup2.select("#gridItemRoot")
```

```python
data2= []
for item in books2:
    
    try:
        title = item.find('div', class_ ="_cDEzb_p13n-sc-css-line-clamp-1_1Fn1y").text
    except Exception as e:
        title = np.nan
    try:
        author = item.find('a', class_ ="a-size-small a-link-child").text
    except Exception as e:
        author = np.nan
        
    book_format = item.find('span', class_ ="a-size-small a-color-secondary a-text-normal").text
    
    review = item.select(".a-size-small")[2].text
    
    rating = item.find('span', class_="a-icon-alt").text.split(' ')[0]
    
    try:
        price = item.find('span', class_ ="_cDEzb_p13n-sc-price_3mJ9Z").text[1:]
    except Exception as e:
        price = np.nan
        
    try:
        year = item.find('h1',class_="a-size-large a-spacing-medium a-text-bold").text[-1]
    except Exception as e:
        year = '2016'
        
    data2.append([title, author, book_format, review, rating, price, year])
```

```python
#put data into a dataframe
book1 = pd.DataFrame(data, columns = ['title','author','book_format','review','rating','price','year'])
```

```python
#put data into a dataframe
book2 = pd.DataFrame(data2, columns = ['title','author','book_format','review','rating','price','year'])
```

```python
#merge both dataframes
book = pd.concat([book1, book2], ignore_index=True)
```
